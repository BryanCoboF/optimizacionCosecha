MODELO DE OPTIMIZACION

```python
#Definicion del modelo
optCosecha = pulp.LpProblem('OptimizacionCosecha',pulp.LpMinimize) #Inicialmente busca minimizar el area de cosecha para obtener el volumen necesario para Enero

# VARIABLE DE DECISION (AREA A COSECHAR)
area_a_cosechar_Ene = pulp.LpVariable.dict("VariableAreaCosecha",colSitioLote, lowBound=0, cat='Continuous')

#print(area_a_cosechar_Ene.values())

#FUNCION OBJETIVO
optCosecha += sum([rend_EneAux[i]*area_a_cosechar_Ene[i] for i in colSitioLote]) # Funcion objetivo obtiene volumen con las areas definidas

# RESTRICCIONES
#Volumen deseado por mes
optCosecha += sum([rend_EneAux[i]*area_a_cosechar_Ene[i] for i in colSitioLote]) >= vecVolminDeseado[0] #Volumen deseado para Enero


for i in area_a_cosechar_Ene:
  #Restringe que el area a cosechar resultante no sea mayor al area de cada lote
  optCosecha += area_a_cosechar_Ene[i]<= areaMax[i]


#Resolucion
optCosecha.solve()


a=1
resAreaCosecha = []
for areaEneCosecha in colSitioLote:

  resAreaCosecha.append(area_a_cosechar_Ene[areaEneCosecha].value())

  print("Area a cosechar: ", a,  areaEneCosecha, "- >", area_a_cosechar_Ene[areaEneCosecha].value())
  a=a+1

res = sum([rend_Ene[i]*area_a_cosechar_Ene[i].value() for i in colSitioLote]) #
print(res)


```
